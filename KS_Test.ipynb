```python
# import packages
import numpy as np
import pandas as pd
import seaborn as sns

# scipy pacage for statistics for using KS Test:
from scipy import stats 
import matplotlib.pyplot as plt

%matplotlib inline

#generate a random variable X

x = stats.norm.rvs(size=1000)

# Assuming the Null HYpothesis (Ho) and Evaluating for Result of 1000 random Variable
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x,"norm")
```

#### Observation :
1. The pvalue is quite high and the stats can be difficult for each random variable.

```python
# y - Continous Uniform Distribution (0,1)

y = np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, "norm")
```

#### Observation :
1.The pvalue is null and the stats is not at all Gaussian
